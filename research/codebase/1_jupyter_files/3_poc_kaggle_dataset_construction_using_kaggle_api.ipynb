### Step 4- Format the Response

Organize and format the obtained response, constructing a Pandas dataframe encompassing all relevant metadata.

By diligently following these systematic steps, users can effectively harness the capabilities of Kaggle APIs to construct Kaggle datasets tailored to specific keywords, thereby optimizing the process of data acquisition and compilation.

Let's create a function where user will provide the keyword and output will be Pandas Dataframe.

We will create a function which will take input as "Topic" we want to search and give us the associated datasets in Pandas dataframe.


Here the input is the list of topics that we want the kaggle datasets for and output is the Pandas dataframw with all the relevant datasets for the topics.

```python
import pandas as pd
from kaggle.api.kaggle_api_extended import KaggleApi

# Assuming you have already authenticated the Kaggle API
api = KaggleApi()
api.authenticate()

COLUMNS_IN_DF = ['dataset_name', 'creatorName', 'creatorUrl',
                  'currentVersionNumber', 'description', 'descriptionNullable',
                  'downloadCount', 'files', 'hasCreatorName', 'hasCreatorUrl', 'hasCurrentVersionNumber',
                  'hasDescription', 'hasLicenseName', 'hasOwnerName', 'hasOwnerRef', 'hasSubtitle', 'hasTitle',
                  'hasTotalBytes', 'hasUrl', 'hasUsabilityRating', 'id', 'isFeatured', 'isPrivate', 'kernelCount',
                  'lastUpdated', 'licenseName', 'licenseNameNullable', 'ownerName', 'ownerNameNullable', 'ownerRef',
                  'ownerRefNullable', 'ref', 'size', 'subtitle', 'subtitleNullable', 'tags', 'title', 'titleNullable',
                  'topicCount', 'totalBytes', 'totalBytesNullable', 'url', 'urlNullable', 'usabilityRating',
                  'usabilityRatingNullable', 'versions', 'viewCount', 'voteCount']

def kaggle_datasets_result(topic_keyword):
    # Create an empty DataFrame with the desired columns
    df = pd.DataFrame(columns=COLUMNS_IN_DF)

    # Retrieve the dataset list based on the search keyword
    datasets = api.dataset_list(search=topic_keyword)

    # Create the DataFrame directly from the list of dictionaries
    df = pd.DataFrame(
        [{'dataset_name': dataset.ref, **{col: getattr(dataset, col, None) for col in COLUMNS_IN_DF[1:]}} for dataset in datasets]
    )
    # Display the first few rows of the DataFrame
    return df


```

```python
output_df = kaggle_datasets_result(user_input)
output_df
```

These are some of the parameters that you can pass in api.dataset_list() function.

sort_by: how to sort the result, see valid_dataset_sort_bys for options

size: Deprecated

file_type: the format, see valid_dataset_file_types for string options

license_name: string descriptor for license, see valid_dataset_license_names

tag_ids: tag identifiers to filter the search

search: a search term to use (default is empty string)

user: username to filter the search to

mine: boolean if True, group is changed to "my" to return personal

page: the page to return (default is 1)

max_size: the maximum size of the dataset to return (bytes)

min_size: the minimum size of the dataset to return (bytes)
