#### Initialising the function which takes input as keyword and gives us Pandas DF as output.

```python
import pandas as pd
from kaggle.api.kaggle_api_extended import KaggleApi

# Assuming you have already authenticated the Kaggle API
api = KaggleApi()
api.authenticate()

COLUMNS_IN_DF = ['dataset_name','title','tags','creatorName', 'creatorUrl', 'currentVersionNumber',
       'description', 'subtitle','descriptionNullable', 'downloadCount', 'files',
       'hasCreatorName', 'hasCreatorUrl', 'hasCurrentVersionNumber',
       'hasDescription', 'hasLicenseName', 'hasOwnerName', 'hasOwnerRef',
       'hasSubtitle', 'hasTitle', 'hasTotalBytes', 'hasUrl',
       'hasUsabilityRating', 'id', 'isFeatured', 'isPrivate', 'kernelCount',
       'lastUpdated', 'licenseName', 'licenseNameNullable', 'ownerName',
       'ownerNameNullable', 'ownerRef', 'ownerRefNullable', 'ref', 'size',
       'subtitleNullable',  'titleNullable',
       'topicCount', 'totalBytes', 'totalBytesNullable', 'url', 'urlNullable',
       'usabilityRating', 'usabilityRatingNullable', 'versions', 'viewCount',
       'voteCount']

def kaggle_datasets_result(topic_keyword):
    # Create an empty DataFrame with the desired columns
    df = pd.DataFrame(columns=COLUMNS_IN_DF)

    # Retrieve the dataset list based on the search keyword
    datasets = api.dataset_list(search=topic_keyword)

    # Create the DataFrame directly from the list of dictionaries
    df = pd.DataFrame(
        [{'dataset_name': dataset.ref, **{col: getattr(dataset, col, None) for col in COLUMNS_IN_DF[1:]}} for dataset in datasets]
    )

    df["keyword"] = topic_keyword
    # Display the first few rows of the DataFrame
    return df


```

### 3. Iterate through every keyword and creation of master CSV which contains all the information about Kaggle datasets.

(A) Batch 1

```python
import time
from tqdm import tqdm

# Master DataFrame to store unique records
master_df = pd.DataFrame(columns=COLUMNS_IN_DF + ["keyword"])

start_time = time.time()

# Iterate through keywords
print("Fetching the dataf for all keywords....")
print("-"*200)

for keyword in tqdm(keywords_batch_1):
    # Create DataFrame for the current keyword
    current_df = kaggle_datasets_result(keyword)
    master_df = pd.concat([master_df, current_df], ignore_index=True)

# Save the master DataFrame to a CSV file
master_df.to_csv("master_dataset.csv", index=False)
print("CSV file created successfully.")
end_time = time.time()
print("-"*200)


# Calculate the total time in minutes
total_time_minutes = (end_time - start_time) / 60


# Display the total time taken
print(f"Total time taken: {total_time_minutes:.2f} minutes")
print("-"*200)
```
